```python
import fix_yahoo_finance as yf
import requests
import json
from tqdm import tqdm
from datetime import timedelta, date
```

```python
def get_stock(date):
    stocks = requests.get(
        "https://www.idx.co.id/umbraco/Surface/TradingSummary/GetStockSummary?date=%s&start=0&length=700" % date
    )
    
    return stocks.json() 
```

```python
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

```python
end_date = date(2019, 1, 1)
start_date = date(2018, 1, 1)
for single_date in tqdm(daterange(start_date, end_date), total = (start_date - end_date).days):
    d = single_date.strftime("%Y%m%d")
    data = get_stock(d)
    
    if data['data']:
        with open('stocks/%s.json' % d, 'w') as outfile:
            json.dump(data, outfile)
```
